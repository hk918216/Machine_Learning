```python
import pandas as pd
```

```python
df = pd.read_csv('./data/06_(문제)타이타닉_분류_데이터셋.csv')
df.info()
```

```python
df = df.drop(['deck'], axis=1)
df = df.drop(['who'], axis=1)
```

```python
df = df.dropna(subset=['age'])
df
```

```python
df = df.reset_index(drop=True)
df
```

```python
df['age'] = df['age'].astype(int)
```

```python
# 독 : sex, pclass, age,/ 종 : survived
```

```python
age_list = []

for i in range(0, len(df['age'])):
    if df['age'][i] >= 0 and df['age'][i] <10:
        age_list.append('0')
    elif df['age'][i] >= 10 and df['age'][i] <20:
        age_list.append('1')
    elif df['age'][i] >= 20 and df['age'][i] <30:
        age_list.append('2')
    elif df['age'][i] >= 30 and df['age'][i] <40:
        age_list.append('3')
    elif df['age'][i] >= 40 and df['age'][i] <50:
        age_list.append('4')
    elif df['age'][i] >= 50 and df['age'][i] <60:
        age_list.append('5')
    elif df['age'][i] >= 60 and df['age'][i] <70:
        age_list.append('6')
    elif df['age'][i] >= 70 and df['age'][i] <80:
        age_list.append('7')
    elif df['age'][i] >= 80 and df['age'][i] <90:
        age_list.append('8')
    elif df['age'][i] >= 90 and df['age'][i] <100:
        age_list.append('9')
    else:
        print('ok')
```

```python
df['연령대'] = age_list
df
```

```python
sex_list = []
for i in range(0,len(df['sex'])):
    if df['sex'][i] == 'male' :
        sex_list.append('0')
    else :
        sex_list.append('1')
```

```python
df['성별'] = sex_list
df
```

- 로지스틱회귀분류

```python
data = df[['pclass','연령대','성별']].to_numpy()
target = df['survived'].to_numpy()
```

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(data, target, random_state=42)

train_input.shape
test_input.shape
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()

ss.fit(train_input)

train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)

print(ss.mean_, ss.scale_)
print("-----------------------")
print(train_scaled[ :5])
print("-----------------------")
print(test_scaled[ :5])
```

```python
### k-최근접이웃분류 모델로 테스트
from sklearn.neighbors import KNeighborsClassifier

kn =KNeighborsClassifier()

kn.n_neighbors=10
kn.fit(train_scaled, train_target)
print(kn.score(test_scaled, test_target))
print(kn.score(train_scaled, train_target))
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(train_scaled, train_target)

print(lr.score(train_scaled, train_target))
print(lr.score(test_scaled, test_target))
```

- sgd

```python
from sklearn.linear_model import SGDClassifier

# 클래스(모델) 생성
sc = SGDClassifier(loss='log',max_iter=10, random_state=42)
sc.fit(train_scaled, train_target)

print(sc.score(train_scaled, train_target))
print(sc.score(test_scaled, test_target))
```

```python
sc.partial_fit(train_scaled, train_target)
print(sc.score(train_scaled, train_target))
print(sc.score(test_scaled, test_target))
```

```python
# 에포크 반복학습
import numpy as np

sc = SGDClassifier(loss = 'log', random_state=42)

train_score = []
test_score = []

# 범주 고유값
classes = np.unique(train_target)

# 반복을 300회 이상으로 테스트하여 정확도를 리스트에 저장
for _ in range(0, 300):
    sc.partial_fit(train_scaled,  train_target, classes = classes)
    
    train_score.append(sc.score(train_scaled, train_target))
    test_score.append(sc.score(test_scaled, test_target))
```

```python
import matplotlib.pyplot as plt

plt.plot(train_score, label='train_score')
plt.plot(test_score, label='test_score')
plt.legend()
plt.show()
```

```python
### 최적의 에포크값 이용해 처음부터 훈련 시작
### 사용 클래스(모델) : SGDClassifier

from sklearn.linear_model import SGDClassifier

# 클래스(모델) 생성
# tol = None : 스스로 훈련이 끝나도 멈추지 말고 지정값까지 돌게 함
sc = SGDClassifier(loss='log',max_iter=50, tol=None, random_state=42)
sc.fit(train_scaled, train_target)
print(sc.score(train_scaled, train_target))
print(sc.score(test_scaled, test_target))
```

- 독립변수 추가

```python
df_total = df[['survived', 'pclass', '연령대', '성별','sibsp','parch','embarked']]
df_total_2 = pd.get_dummies(df_total['embarked'])
df_total = pd.concat([df_total_2, df_total], axis=1)
df_total
```

```python
df_total = df_total.dropna(subset=['embarked'])
df_total = df_total.reset_index(drop=True)
df_total
```

```python
data = df_total[['pclass','연령대','성별','C','Q','S','sibsp','parch']].to_numpy()
target = df_total['survived'].to_numpy()
```

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(data, target, random_state=42)

train_input.shape
test_input.shape
```

```python
### k-최근접이웃분류 모델로 테스트
from sklearn.neighbors import KNeighborsClassifier

kn =KNeighborsClassifier()

kn.n_neighbors=3
kn.fit(train_input, train_target)
print(kn.score(train_input, train_target))
print(kn.score(test_input, test_target))
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(train_input, train_target)

print(lr.score(train_input, train_target))
print(lr.score(test_input, test_target))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_validate
import numpy as np

lr = LogisticRegression()

scores = cross_validate(lr, train_input, train_target, return_train_score=True, n_jobs=-1)

# 최종 훈련평가 결과 및 검증결과
print("훈련 = ", np.mean(scores["train_score"]))
print("테스트 = ", np.mean(scores["test_score"]))
```
